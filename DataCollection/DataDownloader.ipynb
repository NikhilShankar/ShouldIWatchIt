```python
!pip install youtube-comment-downloader
!pip install nltk
!pip install seaborn

import os
import pandas as pd
from youtube_comment_downloader import YoutubeCommentDownloader, SORT_BY_POPULAR
from datetime import datetime
 
# Path to the input CSV file
input_csv_path = "../DataCollection/MovieLists/Real_Hollywood_Movies_2021_2024.csv"
 
# Read the input CSV file
data = pd.read_csv(input_csv_path)
 
# Initialize the YouTube comment downloader
downloader = YoutubeCommentDownloader()
 
# Create the output folder if it doesn't exist
output_folder = "datastore"
os.makedirs(output_folder, exist_ok=True)
 
# Iterate through the movies in the CSV file
for index, row in data.iterrows():
    movie_name = row['Movie Name']
    imdb_rating = row['IMDb Rating']
    youtube_link = row['YouTube Trailer Link']
 
    try:
        # Download comments from the YouTube link
        comments = downloader.get_comments_from_url(youtube_link, sort_by=SORT_BY_POPULAR)
 
        # Create a DataFrame for the comments
        df = pd.DataFrame(comments)
        df['Movie Name'] = movie_name
        df['IMDb Rating'] = imdb_rating
        df.rename(columns={'text': 'Comment'}, inplace=True)
 
        # Save comments to a CSV file
        timestamp = datetime.now().strftime('%Y-%m-%d_%H-%M-%S')
        output_file = os.path.join(output_folder, f"{movie_name}-{timestamp}.csv")
        df.to_csv(output_file, index=False)
 
        print(f"Successfully saved comments for {movie_name} to {output_file}")
    except Exception as e:
        print(f"Failed to download comments for {movie_name}: {e}")
 
print("All comments have been processed.")
```
